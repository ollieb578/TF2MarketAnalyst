```python
import pandas as pd
import matplotlib as plt
import numpy as np

from statistics import mean
```

```python
f = "masterlist.csv"
```

```python
df = pd.read_csv(f)
print(df)
```

```python
df = df.drop(df[df['pricegap']==df['pricegap'].max()].index)
df = df.drop(df[df['pricegap']==df['pricegap'].max()].index)

df = df.drop(df[df['vol']==10000000].index)
```

```python
df = df.drop(df[df.pricegap > 3 * mean(df.pricegap)].index)
df = df.drop(df[df.vol > 10 * mean(df.vol)].index)
```

```python
df.plot(kind='scatter', x='pricegap', y='vol')
```
